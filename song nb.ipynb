# FINAL RECOMMENDER
## USES A PIVOT TABLE WITH SONG_ID VS USER_ID PLAY COUNTS
#### CREATED A GET SONG NAME AND ARTIST NAME HELPER FUNCTION, AND A EUCLIDEAN DISTANCE FUNCTION TO CALCULATE THE TOTAL DISTANCE OF ONE USER_ID'S PLAYCOUNT COLUMN COMPARED TO ANOTHER USER'S PLAYCOUNT COLUMN
#### CALCULATED ALL THE DISTANCES FOR ALL IDS NOT EQUAL TO THE USER_ID, THEN SORTED THE DICTIONARY BY ASCENDING DISTANCE
#### ADDED ALL (AT MOST) TOP 10 SONGS THAT WEREN'T IN THE TOP 5 SONGS PLAYED FOR THE CURRENT ID FOR ALL USER_IDS WITH THE LOWEST DISTANCE(S) IN A DICTIONARY AND THEN SORTED BY PLAY_COUNTS DESCENDING AND RECOMMEND SONGS IN THAT ORDER

```python
#find most similar user. by computing distances between playcount column for specified user to all other users
import random
#combineDups = joined[['userID', 'song_id', 'play_count']].groupby(['userID', 'song_id']).agg(sum)
#combineDups = combineDups.reset_index()
#pivoted = combineDups.pivot(index = 'song_id', columns = 'userID', values = 'play_count').fillna(0)
def get_song_name(song_id):
    return joined[joined['song_id'] == song_id].iloc[0]['title']
def get_artist_name(song_id):
    return joined[joined['song_id'] == song_id].iloc[0]['artist_name']
def euclidean_distance(ID, user_id):
    """
    A simple euclidean distance function
    """
    inner_value = 0
    ID_vals = pivoted[ID].tolist()
    user_id_vals = pivoted[user_id].values.tolist()
    for k in range(len(ID_vals)):
        inner_value += (ID_vals[k] - user_id_vals[k]) ** 2
    return np.sqrt(inner_value)

def rec(user_id=np.random.choice(np.array(pivoted.columns), 1)[0], num=5):
    #SORT BY DISTANCE TO EACH OTHER USER ID
    cols = list(pivoted.columns)[1:]
    d = {}
    for ID in cols:
        if ID == user_id:
            continue
        d[ID] = euclidean_distance(ID, user_id)
    sort = {k: v for k, v in sorted(d.items(), key=lambda item: item[1])}
    
    #RECOMMEND HIGHEST PLAYCOUNT SONGS FOR THE OTHER USER
    #while length, sorted by values - choose all users with same euclidean distance & choose an id randomly
    same_euclid_ids = [k for k,v in sort.items() if v == sort[list(sort.keys())[0]]]
    #rand_id = random.sample(same_euclid_ids, k=1)[0]
    #print(rand_id)


    user_id_top_songs = joined[joined['userID'] == user_id].sort_values(by=['play_count'], ascending=False)['song_id'].values
    top_songs = user_id_top_songs[:5]
    print("FOR USER", user_id, "whose most listened song is ", get_song_name(user_id_top_songs[0]), "by", get_artist_name(user_id_top_songs[0]))
    #get top 10 songs for all "same_euclid_ids" IF exists, then sort dictionary by play_count and list (keep visited dict)
    play_counts = {}
    for rand_id in same_euclid_ids:
        rand_id_top_songs = joined[joined['userID'] == rand_id].sort_values(by=['play_count'], ascending=False)['song_id'].values
        if len(rand_id_top_songs) >= 10:
            rand_id_top_songs = rand_id_top_songs[:10]  
        play = joined[joined['userID'] == rand_id].sort_values(by=['play_count'], ascending=False)['play_count'].values
        j = 0
        for song in rand_id_top_songs:
            if song in top_songs:
                j+=1
                continue
            play_counts[song] = play[j]
            j += 1
    #sort play_counts
    
    play_counts = {k: v for k, v in sorted(play_counts.items(), key=lambda item: item[1], reverse=True)}
    keys = list(play_counts.keys())
    if len(keys) == 0:
        print("no further recommendations available, sorry")

    elif num >= len(play_counts):
        for i in range(len(play_counts)):    
            print(get_song_name(keys[i]), "by", get_artist_name(keys[i]))
    else:
        for i in range(num):    
            print(get_song_name(keys[i]), "by", get_artist_name(keys[i]))
```
